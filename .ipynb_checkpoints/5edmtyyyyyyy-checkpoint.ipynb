```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA

def apply_pca(input_data, desired_explained_variance=0.95):
    """
    Apply PCA to the input data and return the transformed dataset.

    Parameters:
    - input_data: DataFrame, the input dataset.
    - desired_explained_variance: float, the desired cumulative explained variance.

    Returns:
    - transformed_data: DataFrame, the transformed dataset after applying PCA.
    """

    # Initialize the scaler
    scaler = MinMaxScaler()

    # Fit and transform the input data
    input_data_normalized = scaler.fit_transform(input_data)

    # Train dataset
    X_trainp = pd.DataFrame(input_data_normalized, columns=input_data.columns)

    # Performing PCA
    pca = PCA()
    pca.fit(X_trainp)

    # Plot the explained variance ratio
    plt.plot(np.cumsum(pca.explained_variance_ratio_))
    plt.xlabel('Number of Components')
    plt.ylabel('Cumulative Explained Variance')
    plt.title('Explained Variance Ratio by Principal Components')
    plt.show()

    # Choose the number of components based on the desired explained variance
    num_components = np.argmax(np.cumsum(pca.explained_variance_ratio_) >= desired_explained_variance) + 1

    # Project the data onto the retained components
    data_pca = pca.transform(X_trainp)[:, :num_components]

    # The result dataset after applying PCA
    transformed_data = pd.DataFrame(data_pca, columns=[f"PC{i + 1}" for i in range(num_components)])

    return transformed_data

# Example usage:
X_train_transformed = apply_pca(X_train)
X_train_transformed.head()
```

```python
#Test dataset
X_testp = pd.DataFrame(X_test_normalized, columns=X_test.columns)

# Performing PCA
pca = PCA()
pca.fit(X_testp)

# Plot the explained variance ratio
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance Ratio by Principal Components')
plt.show()

# Choose the number of components based on the desired explained variance
desired_explained_variance = 0.95 
num_components = 7
# Project the data onto the retained components
data_pca = pca.transform(X_testp)[:, :num_components]

# The result dataset after applying PCA
X_Test = pd.DataFrame(data_pca, columns=[f"PC{i+1}" for i in range(num_components)])
X_Test.head()
```

```python
from sklearn.mixture import GaussianMixture
import numpy as np

class FGMPM:
    def __init__(self, n_components=3, random_state=42):
        self.n_components = n_components
        self.random_state = random_state
        self.models = []

    def fit(self, X_normal):
        self.models = []
        for feature in X_normal.columns:
            model = GaussianMixture(n_components=self.n_components, random_state=self.random_state)
            model.fit(X_normal[feature].values.reshape(-1, 1))
            self.models.append(model)

    def transform(self, X):
        if not self.models:
            raise ValueError("Fit the model first using the 'fit' method.")

        transformed_features = []
        for i, feature in enumerate(X.columns):
            if i < len(self.models):
                model = self.models[i]
                transformed_feature = model.score_samples(X[feature].values.reshape(-1, 1))
                transformed_features.append(transformed_feature)

        transformed_data = np.vstack(transformed_features).T
        return transformed_data

```

```python
from sklearn.cluster import KMeans


class KMeansDistance:
    def __init__(self, n_clusters=2, random_state=42):
        self.n_clusters = n_clusters
        self.random_state = random_state
        self.kmeans = None

    def fit(self, X_normal):
        self.kmeans = KMeans(n_clusters=self.n_clusters, random_state=self.random_state)
        self.kmeans.fit(X_normal)

    def predict_distance(self, X):
        if not self.kmeans:
            raise ValueError("Fit the model first using the 'fit' method.")

        distances = np.min(self.kmeans.transform(X), axis=1)
        return distances


km_d_model = KMeansDistance(n_clusters=2, random_state=42)
km_d_model.fit(X_train[y_train == b'normal'])  

# Predict distances for the entire dataset (X_train and X_test)
d_raw_kmd_distances = km_d_model.predict_distance(X_train)
d_test_kmd_distances = km_d_model.predict_distance(X_test)
```
