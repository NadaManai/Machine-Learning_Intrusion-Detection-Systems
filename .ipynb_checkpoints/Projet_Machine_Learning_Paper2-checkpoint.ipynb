```python
col_names = ["duration", "protocol_type", "service", "flag", "src_bytes", "dst_bytes", "land",
            "wrong_fragment", "urgent", "hot", "num_failed_logins", "logged_in", "num_compromised",
            "root_shell", "su_attempted", "num_root", "num_file_creations", "num_shells", "num_access_files",
            "num_outbound_cmds", "is_host_login", "is_guest_login", "count", "srv_count", "serror_rate", "srv_serror_rate", "rerror_rate", "srv_rerror_rate", "same_srv_rate", "diff_srv_rate",
            "srv_diff_host_rate", "dst_host_count", "dst_host_srv_count", "dst_host_same_srv_rate", "dst_host_diff_srv_rate", "dst_host_same_src_port_rate", "dst_host_srv_diff_host_rate", "dst_host_serror_rate", "dst_host_srv_serror_rate",
            "dst_host_rerror_rate", "dst_host_srv_rerror_rate", "attack", "last_flag"]
```

```python
#Import the train dataset
dataset_train=pd.read_csv('KDDTrain+.txt',header=None, names = col_names)
dataset_train = dataset_train.iloc[:,:-1]
dataset_train.head()
```

```python
#Import the test dataset
dataset_test=pd.read_csv('KDDTest+.txt',header=None, names = col_names)
dataset_test = dataset_test.iloc[:,:-1]
```

```python
# Remove 'service' and 'flag' columns
dataset_train = dataset_train.drop(['service', 'flag'], axis=1)
dataset_test = dataset_test.drop(['service', 'flag'], axis=1)
```

```python
# Keep only rows with 'protocol_type' as 'tcp'
dataset_train = dataset_train[dataset_train['protocol_type'] == 'tcp']
dataset_test = dataset_test[dataset_test['protocol_type'] == 'tcp']
```

```python
categorical_columns = ["protocol_type"]
encoder = OneHotEncoder(sparse=False, drop='first')
encoded_train = pd.DataFrame(encoder.fit_transform(dataset_train[categorical_columns]))
encoded_test = pd.DataFrame(encoder.transform(dataset_test[categorical_columns]))
```

```python
# Concatenate encoded features with the original dataset
dataset_train = pd.concat([dataset_train.reset_index(drop=True), encoded_train], axis=1)
dataset_test = pd.concat([dataset_test.reset_index(drop=True), encoded_test], axis=1)

# Drop the original categorical columns
dataset_train = dataset_train.drop(categorical_columns, axis=1)
dataset_test = dataset_test.drop(categorical_columns, axis=1)
```

```python
# Define the features and target variable
X_train = dataset_train.drop('attack', axis=1)
y_train = dataset_train['attack']
X_test = dataset_test.drop('attack', axis=1)
y_test = dataset_test['attack']
```

```python
# Apply feature selection using RFE
model = RandomForestClassifier()
n_features_to_select = 24
selector = RFE(model, n_features_to_select=n_features_to_select)
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)

# Get the selected features
selected_features = X_train.columns[selector.support_]

# Update the datasets with selected features
dataset_train_selected = pd.DataFrame(X_train_selected, columns=selected_features)
dataset_train_selected['attack'] = y_train  # Use y_train instead of Y_train

# Display the updated dataset information
print("\nTrain Dataset after Feature Selection:")
print(dataset_train_selected.info())

# Display the number of normal and attack records in the updated train dataset
print("\nTrain Dataset after Feature Selection:")
print("Normal records:", dataset_train_selected[dataset_train_selected['attack'] == 'normal'].shape[0])
print("Attack records:", dataset_train_selected[dataset_train_selected['attack'] != 'normal'].shape[0])
```

```python
# Update the test dataset with selected features
dataset_test_selected = pd.DataFrame(X_test_selected, columns=selected_features)
dataset_test_selected['attack'] = y_test  # Use y_test instead of Y_test

# Display the updated test dataset information
print("\nTest Dataset after Feature Selection:")
print(dataset_test_selected.info())

# Display the number of normal and attack records in the updated test dataset
print("\nTest Dataset after Feature Selection:")
print("Normal records:", dataset_test_selected[dataset_test_selected['attack'] == 'normal'].shape[0])
print("Attack records:", dataset_test_selected[dataset_test_selected['attack'] != 'normal'].shape[0])
```
