```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import arff
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn import preprocessing
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore") 
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn import preprocessing
```

```python
#Import the train dataset
file_path = 'NSL-KDD/KDDTrain+.arff'  # Change the file extension to .arff
data, meta = arff.loadarff(file_path)

# Convert the ARFF data to a Pandas DataFrame
dataset_train = pd.DataFrame(data)

# Display the first few rows of the DataFrame
dataset_train.head()
```

```python
#Import the test dataset
file_path = 'NSL-KDD/KDDTest+.arff'  # Change the file extension to .arff
data1, meta1 = arff.loadarff(file_path)

# Convert the ARFF data to a Pandas DataFrame
dataset_test = pd.DataFrame(data1)

# Display the first few rows of the DataFrame
dataset_test.head()
```

#Remove service, flag and take tcp

```python
dataset_train_tcp = dataset_train[dataset_train['protocol_type'] == b'tcp']
dataset_test_tcp = dataset_test[dataset_test['protocol_type'] == b'tcp']

dataset_test_tcp

#drop service and flag
dataset_train_encoded = pd.get_dummies(dataset_train_tcp, columns=['protocol_type'], drop_first=True)
dataset_test_encoded = pd.get_dummies(dataset_test_tcp, columns=['protocol_type'], drop_first=True)

dataset_train_encoded = dataset_train_encoded.drop(['service', 'flag'], axis=1)
dataset_test_encoded = dataset_test_encoded.drop(['service', 'flag'], axis=1)

num_features_train = dataset_train_encoded.shape[1]
print(f"Number of features in the training dataset: {num_features_train}")
```

#Colloration Matrix

```python
# Exploration des donnÃ©es (statistiques et visuelles)
import seaborn as sns
corr = dataset_train_encoded.corr()
plt.subplots(figsize=(25,10))
sns.heatmap(corr, xticklabels=corr.columns,yticklabels=corr.columns, annot=True, )
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming dataset_train_encoded is your encoded training dataset
corr = dataset_train_encoded.corr().abs()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Find pairs of highly correlated features
highly_correlated = (corr > 0.8) & (corr < 1.0)

# Identify and remove exactly 14 features
features_to_remove = set()
for col in corr.columns:
    correlated_cols = corr.index[highly_correlated[col]].tolist()
    features_to_remove.update(correlated_cols)

# Sort the features based on their overall correlation strength
sorted_features = sorted(features_to_remove, key=lambda x: corr.loc[x, x].sum(), reverse=True)[:15]

# Drop the identified features from the dataset
dataset_train_filtered = dataset_train_encoded.drop(columns=sorted_features)
dataset_test_filtered = dataset_test_encoded.drop(columns=sorted_features)

# Display the remaining features
print("\nRemaining Features:")
print(dataset_train_filtered.columns)
print(dataset_train_filtered.shape[1])

# Visualize the correlation matrix of the remaining features
plt.subplots(figsize=(25, 10))
sns.heatmap(dataset_train_filtered.corr(), xticklabels=dataset_train_filtered.columns, yticklabels=dataset_train_filtered.columns, annot=True)
plt.show()
```

```python
# Check for missing values in the training dataset
print("\nMissing Values in Training Dataset:")
print(dataset_train_filtered.isnull().sum())

# Check for missing values in the test dataset
print("\nMissing Values in Test Dataset:")
print(dataset_test.isnull().sum())

# Explore the distribution of the target variable in the training dataset
print("\nDistribution of 'attack' in Training Dataset:")
print(dataset_train_filtered['class'].value_counts())
```

#OUTLIERS

```python
#Outliers for train set
import seaborn as sns
plt.figure(figsize=(15, 8))
sns.boxplot(data=dataset_train_filtered)
plt.title("Box Plots for Outlier Detection (Training Dataset)")
plt.show()

# Assuming dataset_train is your training dataset
# Calculate and display the IQR for each numeric column
numeric_cols = dataset_train_filtered.select_dtypes(include=np.number).columns
Q1 = dataset_train_filtered[numeric_cols].quantile(0.25)
Q3 = dataset_train_filtered[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers using the IQR method for numeric columns
outliers_numeric = ((dataset_train_filtered[numeric_cols] < (Q1 - 1.5 * IQR)) | (dataset_train_filtered[numeric_cols] > (Q3 + 1.5 * IQR)))

# Display the number of outliers in each numeric column
print("\nNumber of outliers in each numeric column:")
print(outliers_numeric.sum())
```

```python
#Outliers for test set
import seaborn as sns
plt.figure(figsize=(20, 15))
sns.boxplot(data=dataset_test_filtered)
plt.title("Box Plots for Outlier Detection (Testing Dataset)")
plt.show()

# Assuming dataset_train is your training dataset
# Calculate and display the IQR for each numeric column
numeric_cols = dataset_test_filtered.select_dtypes(include=np.number).columns
Q1 = dataset_test_filtered[numeric_cols].quantile(0.25)
Q3 = dataset_test_filtered[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers using the IQR method for numeric columns
outliers_numeric = ((dataset_test_filtered[numeric_cols] < (Q1 - 1.5 * IQR)) | (dataset_test_filtered[numeric_cols] > (Q3 + 1.5 * IQR)))

# Display the number of outliers in each numeric column
print("\nNumber of outliers in each numeric column:")
print(outliers_numeric.sum())
```

```python
from sklearn.impute import SimpleImputer

# Assuming dataset_train_filtered is your filtered training dataset

# Separate numeric and non-numeric columns
numeric_cols = dataset_train_filtered.select_dtypes(include=np.number).columns
non_numeric_cols = dataset_train_filtered.select_dtypes(exclude=np.number).columns

# Calculate and display the IQR for each numeric column
Q1 = dataset_train_filtered[numeric_cols].quantile(0.25)
Q3 = dataset_train_filtered[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers using the IQR method for numeric columns
outliers_numeric = ((dataset_train_filtered[numeric_cols] < (Q1 - 1.5 * IQR)) | (dataset_train_filtered[numeric_cols] > (Q3 + 1.5 * IQR)))

# Create a copy of the dataset to avoid modifying the original
dataset_train_imputed = dataset_train_filtered.copy()

# Set outliers to NA for numeric columns
dataset_train_imputed[numeric_cols][outliers_numeric] = np.nan

# Impute missing values for numeric columns using SimpleImputer with 'mean' strategy
imputer_numeric = SimpleImputer(strategy='mean')
dataset_train_imputed[numeric_cols] = imputer_numeric.fit_transform(dataset_train_imputed[numeric_cols])

# Impute missing values for non-numeric columns using SimpleImputer with 'most_frequent' strategy
imputer_non_numeric = SimpleImputer(strategy='most_frequent')
dataset_train_imputed[non_numeric_cols] = imputer_non_numeric.fit_transform(dataset_train_imputed[non_numeric_cols])

# Display the number of imputed values in each column
print("\nNumber of imputed values in each column:")
print(dataset_train_imputed.isnull().sum())
```
